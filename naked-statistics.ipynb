# Introduction

***It is easy to lie with statistics, but it's hard to tell the truth without them!***

Statistics can be really interesting, and most of it isn't that difficult. I'm sure our professor must have taught this concept under which *the sum of infinite series (1 + 1/2 + 1/4 + 1/8 + ...) converges to a finite number.* It is so difficult to digest that how can something that is infinite add up to something that is finite? 

Now take a moment and imagine you have positioned yourself exactly 2 feet from a wall. Now move half the distance to the wall (1 foot), so that you're left standing 1 foot away. From one foot away, move half the distance to the wall once again (6 inches, or 1/2 a foot). And from 6 inches away , do it again (move 3 inches, or 1/4 of a foot). the do it again (move 1(1/2),  or 1/8 of a foot). And so on.

You will gradually get pretty darn close to the wall. For eg. when you are 1/1024th of an inch from the wall, you'll move half the distance, or another 1/2048th of an inch. But you'll never hit the wall, because by definition each move takes you only half the remaining distance. In other words, you will get close to the wall but never hit it. If we measure the moves, the series can be described as 1+ 1/2 + 1/4 + 1/8 ... 

Even though you will continue moving forever - the total distance can never be more than 2 feet, which is the starting distance from the wall. For mathematical purposes, the total distance travelled can be approximated as 2 feet, which is very handy for computational purposes. A mathematician would saay that the sum of this infinite series 1 ft + 1/2 ft + 1/4 ft + 1/8 ft ... converges to 2 feet. 

Now that we know the application, *the sum of infinite series (1 + 1/2 + 1/4 + 1/8 + ...) converges to a finite number*, this sentence makes sense now!
